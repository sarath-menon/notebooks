```julia
using ControlSystems
using Plots
```

# Transfer function

### Create system

```julia
P = tf([1.0],[1,1])
```

### Add a feedback connection

```julia
C = 2
fb_sys = feedback(P*C)
```

### Adding delay

```julia
# create delay of 0.2 s
time_delay = delay(0.2);

# Input delay
fb_sys_delayed = fb_sys*time_delay

# Output delay
fb_sys_delayed = time_delay*fb_sys
```

### Plotting

```julia
bodeplot(fb_sys)
```

# State-Space

```julia
A,B,C = [-5 0; 0 -5],[2; 2],[3 3]
```

### Create continuous time system

```julia
sys_ct = ss(A,B,C,0);
```

### Create discrete time system

```julia
# sampling tme
Ts = 0.1;

sys_dt = ss(A,B,C,0, Ts);
```

### State space to transfer function

```julia
tf(sys_ct)
```
