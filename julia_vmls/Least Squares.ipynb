```julia
using LinearAlgebra
using StatsBase
using Plots
```

## Problem 1: Advertaisement allocation

### Define problem

```julia
R = [ 0.97 1.86 0.41;
    1.23 2.18 0.53;
    0.80 1.24 0.62;
    1.29 0.98 0.51;
    1.10 1.23 0.69;
    0.67 0.34 0.54;
    0.87 0.26 0.62;
    1.10 0.16 0.48;
    1.92 0.22 0.71;
    1.29 0.12 0.62];

m, n = size(R); 

V_des = vdes = 1e3 * ones(m);
```

### Solution

```julia
# solution via backslash operator
S = R \ V_des;

# Compute RMS error
V_actual = R*S;
rms_error = sqrt(msd(V_actual, V_des));

print("Least Squares Solution:");
display(S);
println("RMS Error:", rms_error);

# Plot the solution
bar(V_actual,  label="V_actual")
# plot desired solution to compare results
plot!(V_des, label="V_des", lw=3)
```
