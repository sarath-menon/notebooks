## Parametrizing curves

```julia

function get_cartesian(;r::Vector{Float64}, θ)
    # get cartesian coordinates 
    x = r .* cos.(θ)
    y = r .* sin.(θ)
            
    return (x,y)
end
        
```

### Example 1: rhodenea

```julia
let 
# constants
a, k = 4.0, 5.0

# curve function
r(θ) = a * sin(k*θ)
    
θ_vec = range(0, pi, length=100)
r_vec = r.(θ_vec )

(x,y) = get_cartesian(r=r_vec, θ=θ_vec)

# plotting 
fig = lines(x, y)
display(fig)

end
```

## Area under polar graph

Theory: https://www.youtube.com/watch?v=qVn_Lfec-Ac

```julia
let
@syms θ

# curve function: cardioid
r(θ) = 2 * (1 + cos(θ))

# get cartesian coordinates
θ_vec = range(0, 2*π, length=100)
r_vec = r.(θ_vec)
(x,y) = get_cartesian(r=r_vec, θ=θ_vec)

# plotting 
fig = lines(x, y)
display(fig)

# Integration
θ_low = 0
θ_high = 2*π
    
area = (1//2) * integrate(r(θ)^2, (θ, θ_low, θ_high))

end
```

## Arc length

```julia
let

@syms θ

# curve function: equiangular spiral
a, b = 1.0, 1.3
r(θ) = a * exp(θ * cot(b))

# get cartesian coordinates
θ_vec = range(0, 3*π, length=100)
r_vec = r.(θ_vec)
(x,y) = get_cartesian(r=r_vec, θ=θ_vec)

# plotting 
fig = lines(x, y)
display(fig)

# integration
θ_low = 0
θ_high = 2*PI

ds = sqrt(diff(r(θ), θ)^2 + r(θ)^2)    
integrate(ds, (θ, 0, 1))

end
```
