```julia
using GLMakie
using LinearAlgebra
#using Symbolics
using SymPy
using ForwardDiff

GLMakie.activate!(inline=true)
```

```julia
x(s, t) = 3*cos(t) + cos(t)cos(s)
y(s, t) = 3*sin(t) + sin(t)cos(s)
z(s, t) = sin(s)

f(s, t) = [x.(s, t) y.(s, t) z.(s, t)] 
```

## Case 1: s=0

```julia
f_total_vec = Point3f[];

for t in 0:0.1:2*π
    for s in 0:0.1:2*π
        f_vec =  f(0, t)
        push!(f_total_vec, f_vec)
    end
end

fig = lines(f_total_vec)
display(fig)
```

## Case 2: t=0

```julia
f_total_vec = Point3f[];

for t in 0:0.1:2*π
    for s in 0:0.1:2*π
        f_vec =  f(s, 0)
        push!(f_total_vec, f_vec)
    end
end

fig = lines(f_total_vec)
display(fig)
```

## Case 3: Both non-zero

```julia
f_total_vec = Point3f[];

for t in 0:0.1:2*π
    for s in 0:0.1:2*π
        f_vec =  f(s, t)
        push!(f_total_vec, f_vec)
    end
end

fig = lines(f_total_vec)
display(fig)
```
