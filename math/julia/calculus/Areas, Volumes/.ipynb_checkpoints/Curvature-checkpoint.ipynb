### Finding curvature - Differentiation w.r.t arc length 

```julia
let 
    
@syms t::(real,positive) 

D = Differential(t)
    
x(t) = t - sin(t)
y(t) = 1 - cos(t)

s(t) = [x.(t) y.(t)] 

# tangent vector (derivative)
T = D.(s(t)) 

# unit tangent vector (normalized)

T_unit = T ./norm(T)

# differentiate unit tangent vector w.r.t path length(s)
Ds = Differential(t)
Ds.(T_unit) 
    
end
```

# Example 2: Circle

### Symbolics.jl version

```julia
let 
    
@variables t, R

D = Differential(t)
    
x(t) = R*cos(t)
y(t) = R*sin(t)

s(t) = [x.(t) y.(t)] 

# tangent vector (derivative)
T = expand_derivatives.( D.(s(t)) )

# unit tangent vector (normalized)

T_unit = T ./norm(T)


# differentiate unit tangent vector w.r.t path length(s)
Ds = Differential(t)
norm(expand_derivatives.( Ds.(T_unit) ))
    
end
```

### Sympy.jl version

```julia
let 
    
@syms t::(real,positive) R::(real,positive)

# R, t = symbols("R t", positive=true, real=true)

D = Differential(t)
    
x(t) = R*cos(t)
y(t) = R*sin(t)

s(t) = [x.(t) y.(t)] 

# tangent vector (derivative)
T = D.(s(t)) 

# unit tangent vector (normalized)

T_unit = T ./norm(T)

simplify(T_unit)

# differentiate unit tangent vector w.r.t path length(s)
Ds = Differential(t)
norm( Ds.(T_unit) )
    
end
```

# Example 3: Helix

```julia
let 

x(t) = cos(t)
y(t) = sin(t)
z(t) = t/5

s(t) = [x.(t) y.(t) z.(t)] 

t_vec = collect(0:0.1:10)
s_vec = s(t_vec)
    
lines(s_vec)
    
end
```

### Sympy.jl version

```julia
let 
    
@syms t::(real,positive) R::(real,positive)

D = Differential(t)
    
x(t) = cos(t)
y(t) = sin(t)
z(t) = t/5

s(t) = [x.(t) y.(t) z.(t)] 

# tangent vector (derivative)
T = D.(s(t)) 

# unit tangent vector (normalized)

T_unit = T ./norm(T)

simplify(T_unit)

# differentiate unit tangent vector w.r.t path length(s)
Ds = Differential(t)
norm( Ds.(T_unit) )
    
end
```
