```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Consider that we are identfiying the spring constant $k$. The dynamics is descrivbed by $k = \frac{f(t)}{x(t)}$, where $f(t)$ is the applied force and $x(t)$ is the displacement. We are allowed to take 10 measurements. The the true value of $k$ be 1.5

```python
n_samples = 10
k_true = 1.5
```

Step 1: Create a vector of $Y$ fake measurements. We assume that each measurement $\hat{k}$ is i.i.d, and model it as (true value of $k$ + zero mean gaussian noise). Let the gaussian have a standard deviattion of 0.2

```python
noise_gaussian_std_dev = 0.1
noise_gaussian_var = np.sqrt(noise_gaussian_std_dev)
Y = np.random.normal(k_true, noise_gaussian_var , size=(1, n_samples))
Y
```

## Plot gaussian

```python
# Plot between -10 and 10 with .001 steps.
x_axis = np.arange(0, 4, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, k_true, noise_gaussian_std_dev), label='μ:' + str(k_true) + ', σ:' + str(noise_gaussian_std_dev), color='red')

#add legend to plot
plt.legend(title='Parameters')

#add axes labels and a title
plt.ylabel('Density')
plt.xlabel('x')
plt.title('Normal Distributions', fontsize=14)

plt.grid(True)
plt.show()
```
