# Example: find values larger than threshold in sin wave

### Create sin wave

```python
dt  = 0.001
duration = 4
freq = 2
amplitude = 1.0

t = np.arange(0,duration,dt)
signal = amplitude*np.sin(2*np.pi*freq*t)
# plt.plot(t, signal)
# plt.show()
```

### threshold the signal

```python
# set a threshold
threshold = 0.7

is_greater = signal > threshold
# copy the signal explicitly
signal_thresholded = signal.copy()

# set the elements greater than the thresholds to nan
signal_thresholded[~is_greater] = np.nan
# plt.plot(t, signal_copy)
# plt.show()
```

### plot the signal

```python
 # Plot the data
fig, axs = plt.subplots(3,1)

axs[0].plot(t, signal)
axs[0].plot(t, signal, lw=3)
axs[1].plot(t[is_greater], signal[is_greater], '*-')
axs[2].plot(signal[is_greater])

# Format the plot
axs[0].set_ylabel('All data')
axs[0].axhline(threshold, ls='dotted')
axs[0].margins(x=0)
axs[0].set_xticklabels([])

axs[1].set_ylabel('Large data')
axs[1].set_xlabel('Time [s]')
axs[1].margins(x=0)
axs[1].set_xlim(0, duration)
axs[1].set_ylim(-1.05, 1.05)

axs[2].set_ylabel('Large data only')
axs[2].set_xlabel('Points only')

# Group the top two axes, since they have the same x-scale
axs[0].set_position([0.125, 0.75, 0.775, 0.227])
axs[1].set_position([0.125, 0.50, 0.775, 0.227])
axs[2].set_position([0.125, 0.09, 0.775, 0.227])

# Save and show the figure
# out_file = 'FindingEvents.jpg'
# plt.savefig(out_file, dpi=200, quality=90)
# print(f'Image saved to {out_file}')

plt.show()
```
