## Differentiator as low pass filter

The F.T of $f'(x)$ gives $F(X)$ scaled by $ j \omega$ with no change in phase. This corresponds to a high pass filter in the frequency domain. Here, we use the laplace variable $ s = j \omega$ for the bode plot (bug: shows 90 deg change in phase)

```python
s = cs.tf('s')
G  = s
mag, phase, omega = cs.bode(G)
```

# Using scipy

```python
from scipy import signal
import matplotlib.pyplot as plt

sys = signal.TransferFunction([1, 0], [1])

w, mag, phase = signal.bode(sys)

fig_height = 2
fig_width = 6

# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(fig_width)
f.set_figheight(fig_height)

plt.xlabel("Frequency (rad/sec)")
plt.ylabel("Amplitude (dB)")
# Display grid
plt.grid(True, which="both")
plt.semilogx(w, mag)    # Bode magnitude plot

# plotting a line plot after changing it's width and height
f = plt.figure()
f.set_figwidth(fig_width)
f.set_figheight(fig_height)
plt.xlabel("Frequency (rad/sec)")
plt.ylabel("Magnitude (dB)")
plt.grid(True, which="both")
plt.semilogx(w, phase)  # Bode phase plot
plt.show()
```
