```python
import numpy as np
import matplotlib.pylab as plt
from scipy import signal as dsp
from scipy.fft import fft, fftfreq
import plotter
import signals
```

## Lowass filter

```python
numtaps = 1001
f_s = 1000 #sampling frequency in [Hz]
f_c = 50 #cutoff frequency in [Hz]

f_n = f_c / (f_s * 0.5) # cutoff frequency normalized to nyquist frequency

filter_coeff = dsp.firwin(numtaps, f_n, pass_zero="lowpass")

plotter.plot_bode_magn(filter_coeff, f_s)
```

## Highpass filter 

```python
numtaps = 1001
f_s = 1000 #sampling frequency in [Hz]
f_c = 300 #cutoff frequency in [Hz]

f_n = f_c / (f_s * 0.5) # normalized to nyquist frequency

filter_coeff = dsp.firwin(numtaps, f_n, pass_zero="highpass")

plotter.plot_bode_magn(filter_coeff, f_s)
```

## Bandpass

```python
numtaps = 1001
f_s = 1000 #sampling frequency in [Hz]
f_start = 50 #[Hz]
f_stop = 150 #[Hz]

# normalized to nyquist frequency
f_start_n = f_start / (f_s * 0.5) 
f_stop_n = f_stop / (f_s * 0.5) 

filter_coeff = dsp.firwin(numtaps, [f_start_n,f_stop_n], pass_zero="bandpass")

plotter.plot_bode_magn(filter_coeff, f_s)
```
