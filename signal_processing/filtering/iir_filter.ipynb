```python
import numpy as np
import matplotlib.pylab as plt
from scipy import signal as dsp
```

# Butterworth filter 

### Lowpass

```python
f_s = 16000; # sampling frequency in Hz
m = 5; # filter order 
# f_c = np.array([100,200]) # cutoff frequency in Hz

w_n_hz = 100 # point at which gain drops to 1/sqrt(2) that of the -3 dB point
w_n = 2*np.pi*w_n_hz
```

```python
 # design filter 
[b,a] = dsp.butter(m, w_n, btype='lowpass', fs=f_s)

# get frequency response 
w, h = dsp.freqs(b, a)
# w_hz = f_s* w/ (2*np.pi)

# plotting 
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='green') # cutoff frequency
plt.show()
```

### Bandpass

```python
f_s = 1000; # sampling frequency in Hz
m = 5; # filter order 
f_c = np.array([100,200]) # cutoff frequency in Hz

# comoute normalized freuency 
f_n  = 2*(f_c/f_s)
```

```python
 # design filter 
[b,a] = dsp.butter(m, f_n, btype='bandpass')

# get frequency response 
w, h = dsp.freqs(b, a)
w_hz = f_s* w/ (2*np.pi)

# plotting 
plt.plot(w, abs(h))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
# plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(100, color='green') # cutoff frequency
plt.show()
```
