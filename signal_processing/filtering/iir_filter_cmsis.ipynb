```python
import numpy as np
import matplotlib.pylab as plt
from scipy import signal as dsp
import plotter
import signals
```

## Create noisy input signal

```python
duration = 0.2 #s
sampling_rate = 1000 #Hz

f_true = 50 #hz
f_noise = 150 #hz
    
true_signal, t = signals.create_sin_wave(f_true, duration, sampling_rate)
noise, t = signals.create_sin_wave(f_noise, duration, sampling_rate)

noisy_signal = true_signal + noise
noisy_signal_f32 = noisy_signal.astype(np.float32)
noisy_signal_f32 = noisy_signal_f32 / np.max(noisy_signal_f32)

plt.plot(t, noisy_signal_f32)
```

## Chebyshev Type 1

```python
f_s = 1000; # sampling frequency in Hz

rp = 3 # max ripple in dB
m = 4; # filter order 
f_c = 100  # cutoff frequency in Hz

# design filter 
b, a = dsp.cheby1(m, rp=rp, Wn=f_c, fs=f_s, btype='lowpass')

# get frequency response 
w, h = dsp.freqz(b, a)
w_hz = (f_s * w)/ (2*np.pi)

mag = abs(h)

plotter.filter_plotter(w_hz, mag, f_c, amplitude_input_unit='normal')
```

```python
sos = dsp.cheby1(m, rp=rp, Wn=f_c, fs=f_s, btype='lowpass', output='sos')

filtered_signal = dsp.sosfilt(sos, noisy_signal)
# filtered_signal = filtered_signal.astype(np.float32)

fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(t, noisy_signal)
ax2.plot(t, filtered_signal)

ax2.set_title('After Chebyshev 1 filter')
ax2.set_xlabel('Time [seconds]')

plt.tight_layout()
plt.show()
```

# CMSIS DSP implementation

```python
print("No of sections: ", sos.shape[0])
n_sections = sos.shape[0]

# each row of form [b0, b1, b2, a0, a1, a2] corresponds to a 2nd order section 

# Normalize coefficients such that a_0 is 1 and flip the sigs of the a coefficientss
sos_normalized = np.empty(sos.shape)

for (i,section) in enumerate(sos):
    a0 = section[3]
    print("a0"+str(i), ": ", a0)
    # normalize 
    sos_normalized[i,:] =  section / a0
    # flip sign of a 
    sos_normalized[i,3:6] = -sos_normalized[i,3:6]

# delete a0 and reshape to single axis array
sos_normalized_a0_removed = np.delete(sos_normalized, 3, 1)
sos_normalized_a0_removed = sos_normalized_a0_removed .reshape(sos_normalized_a0_removed .size)

print("sos: ", sos)
print("sos normalized: ", sos_normalized)
print("sos normalized , a0 removed : ", sos_normalized_a0_removed)

# scale down by 4 
sos_normalized_a0_removed = sos_normalized_a0_removed/4.0
```

```python
from cmsisdsp import arm_biquad_casd_df1_inst_q31, arm_float_to_q31, arm_biquad_cascade_df1_init_q31, arm_biquad_cascade_df1_q31

# convert input signal to q31 format
noisy_signal_q31 = arm_float_to_q31(noisy_signal_f32)

# instance of floating-point Biquad cascade filter
biquadQ31 = arm_biquad_casd_df1_inst_q31()

# state array to initialize instance
numStages=n_sections
state=np.zeros(numStages*4)

# convert filter coeff to q31 format
coeff_q31 = arm_float_to_q31(sos_normalized_a0_removed)

postshift = 2
arm_biquad_cascade_df1_init_q31(biquadQ31,numStages,coeff_q31,state,postshift)

#  apply filter 
filtered_signal_cmsis = arm_biquad_cascade_df1_q31(biquadQ31, noisy_signal_q31)
```
