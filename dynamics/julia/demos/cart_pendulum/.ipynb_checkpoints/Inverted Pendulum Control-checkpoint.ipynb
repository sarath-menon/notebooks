```julia
using DifferentialEquations, ModelingToolkit
# using GLMakie
using ControlSystemsBase, RobustAndOptimalControl, ForwardDiff
using Plots
```

```julia
function convert_to_numeric(A::Array{Num})
    A_num = [i.val for i in A]
    A_num = convert(Array{Float64}, A_num)
    
    return A_num
end
```

```julia
# define parameters and variables
@parameters ell  γ
@variables t θ(t) y(t) y1(t) [output= true] τ(t)=0 [input = true]
@constants g_=9.81  

# define operators
D = Differential(t)

# dynamics equations
eqs = [
    D(D(θ)) ~ τ + -(g_ / ell) * sin(θ) - γ*D(θ),
    y ~ θ]

# create ODESystem and simplify
@named sys = ODESystem(eqs)
sys = structural_simplify(sys)
```

### Simulation

```julia
# initial conditions
@nonamespace u₀ = [
    sys.θ => π + 0.1,
    sys.θˍt => 0.0]

# parameter values
@nonamespace params = [
    sys.ell => 1.0,
    sys.γ => 0.01]

#Pass to solver
tspan = (0.0, 10.0)
prob = ODEProblem(sys, u₀, tspan, params, jac=true)
sol = solve(prob, Tsit5(), abstol=1e-8, reltol=1e-8);

Plots.plot(sol)
```

### Linearization

```julia
# manual linearization
params =  Dict(ell => 1.0, g_ => 9.81, γ=> 0.01)
init_cond = Dict(θ => π)

# substitute parameters 
eq_subs = substitute(eqs[1],params)

# Calculate A,B matrices
Ac = substitute(Symbolics.jacobian([D(θ), eq_subs.rhs],[θ, D(θ)]), init_cond)
Bc = substitute(Symbolics.jacobian([D(θ), eq_subs.rhs],[τ]), init_cond)
Cc = [1 0]

# Create state space system 
sys_c = ss(convert_to_numeric(Ac), convert_to_numeric(Bc), Cc, 0)
```

### PID controller using MTK

```julia
function plant(; name)
    # define parameters and variables
    @parameters ell  γ
    @variables t θ(t) y(t) y1(t) u(t) [input = true]
    @constants g_ = 9.81  
    
     # define operators
    D = Differential(t)
    
    y1 = D(θ)

    # dynamics equations
    eqs = [
        D(D(θ)) ~ u + -(g_ / ell) * sin(θ) - γ*D(θ),
        y ~ θ,
    ]

    # create ODESystem and simplify
    @named sys = ODESystem(eqs, t; name = name)
end


function pid_ctrl(;kp,ki,kd, name)
    @variables t u(t)=0 e_accum(t)=0 y(t)=0 [input=true] r(t)=0 [input=true] e(t)=0
    @parameters kp=kp ki=ki kd=kd
    
     # define operators
    D = Differential(t)

    eqs = [
        e ~ r - y,       # error
        D(e_accum) ~ e , # accumulated error 
        u ~ kp*e + ki*e_accum+ kd*D(e)
    ]
    
    ODESystem(eqs, t; name = name)
end


function step(;height, name)
    @variables y(t) [output = true] 
    @parameters height=height
    eqs = [
        y ~ height
    ]
    ODESystem(eqs, t; name = name)
end
```

```julia
using ModelingToolkit: connect

function SystemModel(p,c,r)
    eqs = [
        r.y ~ c.r, 
        c.u ~ p.u, # controller output to plant input
        p.y ~ c.y] # plant output to controller input

    @named sys = ODESystem(eqs, t, systems = [c, p, r])
    sys = structural_simplify(sys)
end

@named p = plant()
@named c = pid_ctrl(kp=30, ki=0, kd=3)
@named r = step(height=π)

# create model of whole system
sys = SystemModel(p,c,r);

# initial conditions
@nonamespace u₀ = [
    sys.p₊θ => π - 0.2,
    sys.p₊θˍt => 0.0]

# parameter values
@nonamespace params = [
    sys.p₊ell => 1.0,
    sys.p₊γ => 0.01]

#Pass to solver
tspan = (0.0, 10.0)
prob = ODEProblem(sys, u₀, tspan, params, jac=true)
sol = solve(prob, Tsit5(), abstol=1e-8, reltol=1e-8);

pendulum_plot(sol)
```

### Plotting

```julia
using GLMakie

function pendulum_plot(sol::ODESolution)
    fig = Figure(resolution=(800, 400))

    position_ax = Axis(fig[1, 1], title="Angle")
    velocity_ax = Axis(fig[2, 1], title="Angular Velocity")

    lines!(position_ax, sol.t, sol[sys.p₊θ])
    lines!(position_ax, sol.t, ones(length(sol.t)) * π)
    lines!(velocity_ax, sol.t, sol[sys.p₊θˍt])

    position_ax.ylabel = "angle [rad]"
    velocity_ax.ylabel = "angular velocity [rad/s]"
    velocity_ax.xlabel = "time [s]"

    rowgap!(fig.layout, 5)

    fig
end
```
