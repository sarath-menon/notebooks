```python
from control.matlab import *
import numpy as np
import matplotlib.pyplot as plt
```

# System Definiton - Spring mass Damper

```python
# Parameters defining the system
m = 250           # system mass
k = 40.0            # spring constant
b = 60.0            # damping constant

# System matrices
A = [[0, 1.], [-k/m, -b/m]]
B = [[0], [1/m]]
C = [[1., 0]]
sys = ss(A, B, C, 0)
```

# System Definiton - Pendulum

```python
# Parameters defining the system
g = -9.81         
l = 0.1           # string length
b = 0.2           # damping coefficient

# System matrices
A = [[0, 1.], [g/l, -b/l]]
B = [[0], [1/l]]
C = [[1., 0]]
sys = ss(A, B, C, 0)
```

# Step Response

```python
plt.figure(1)
yout, T = step(sys)
plt.plot(T.T, yout.T)
plt.grid()
plt.show(block=False)
```

# Impulse Response

```python
plt.figure(1)
yout, T = impulse(sys)
plt.plot(T.T, yout.T)
plt.grid()
plt.show(block=False)
```

#  Initial condition sim

```python
plt.figure(1)
X0 = [1., 0.]    # initial condition
yout, T = initial(sys, X0=X0)
plt.plot(T.T, yout.T)
plt.grid()
plt.show(block=False)
```

# Bode Plot

```python
# Bode plot for the system
plt.figure(2)
mag, phase, om = bode(sys, logspace(-2, 2), plot=True)
plt.show(block=False)
```
