```python
from control.matlab import tf, step, ss, bode
import numpy as np
import matplotlib.pyplot as plt
```

# Create transfer function

## method 1: Polynomial form

```python
sys = tf([1], [1,2,3])
sys
```

## method 2: Symbolic form

```python
s = tf('s')
G  = (s + 1)/(s**2 + 2*s + 1)
G
```

# State-space to transfer function

## Step 1: create state-space description (spring-mass damper)

```python
# Parameters defining the system
m = 250           # system mass
k = 40.0            # spring constant
b = 60.0            # damping constant

# System matrices
A = [[0, 1.], [-k/m, -b/m]]
B = [[0], [1/m]]
C = [[1., 0]]
sys_ss = ss(A, B, C, 0)
```

## Step 2: create transfer function representation

```python
sys_tf = tf(sys_ss)
sys_tf
```
